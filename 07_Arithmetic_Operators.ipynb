### 1. Arithmetic Operators
Used for basic mathematical operations.

| Operator | Name | Example | Result |
| :---: | :--- | :--- | :---: |
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division (float) | `5 / 2` | `2.5` |
| `//` | Floor Division (integer) | `5 // 2` | `2` |
| `%` | Modulus (remainder) | `5 % 2` | `1` |
| `**` | Exponentiation (power) | `5 ** 3` | `125` |

---

### Addition (+)

```python
a = 10
b = 5
result = a + b
print(result)
```

```python
x = 3.5
y = 2.7
print(x + y)
```

```python
first_name = "Hasib"
last_name = "Hossain"
full_name = first_name + " " + last_name
print(full_name)  
```

### Subtraction (-)

```python
a = 20
b = 8
result = a - b
print(result)
```

```python
price = 100.50
discount = 25.25
final_price = price - discount
print(f"Final Price: ${final_price}")
```

###  Multiplication (*)

```python
a = 6
b = 7
result = a * b
print(result)
```

```python
length = 5.5
width = 3.2
area = length * width
print(f"Area: {area}")
```

```python
print("Hasib " * 10)
```

### Division (/)

```python
a = 10
b = 3
result = a / b
print(result)
```

### Floor Division (//)
Returns integer part of division (rounds down to nearest whole number).

```python
a = 17
b = 5
result = a // b
print(result)
```

### Modulus (%)
Returns the remainder of division.

```python
a = 17
b = 5
remainder = a % b
print(remainder)
```

### Exponentiation (**)
Raises left operand to the power of right operand.

```python
a = 5
square = a ** 2
print(f"{a}² = {square}")
```

```python
b = 3
cube = b ** 3
print(f"{b}³ = {cube}")
```
