### Conditional Statements

 What Are Conditional Statements?

Conditional statements allow your program to make decisions and execute different code based on certain conditions. They are the **foundation of logic** in programming.

>  **Think of it like:** If it's raining, take an umbrella. Otherwise, go outside.

---

Types of Conditional Statements

| Statement | Purpose | When to Use |
|-----------|---------|-------------|
| `if` | Runs code if condition is `True` | Single condition check |
| `elif` | Short for "else if" – checks another condition | Multiple related conditions |
| `else` | Runs code if all previous conditions are `False` | Fallback/default case |
| Nested `if` | `if` statement inside another `if` | Complex logic with dependencies |

### The `if` Statemen

```python
age = 18

if age >= 18:
    print("You are eligible to vote.")
```

### The if-else Statement

```python
age = 16

if age >= 18:
    print("You can vote.")
else:
    print("You are too young to vote.")
```

```python
number = 7     

if number % 2 == 0:
    print(number, "is an even number.")
else:
    print(number, "is an odd number. ")   
```

### The if-elif-else Statement

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: c")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")                
```

### Nested Conditionals

```python
age = 20
has_license = True

if age >= 18:
    if has_license:
        print("You can drive.")
    else:
        print("You need a driver's license to drive.")
else:
    print("You are too young to drive.")            
```
