> A list is a mutable (changeable), ordered collection that allows duplicate elements.

### Creating lists

```python
# Creating lists
empty_list = []
number = [1, 2, 3, 4, 5]
mixed_data_types = [1, "Hello", 3.4]
fruits = ["Apple", "Banana", "Cherry"]
```

### Accessing Elements

```python
fruits = ["Apple", "Banana", "Cherry"]

print(fruits[0]) # Output: Apple
print(fruits[1]) # Output: Banana
print(fruits[-1]) # Output: Cherry
print(fruits[1:3]) # Output: ['Banana', 'Cherry']
```

### Adding Elements

```python
fruits = ["apple", "banana"]

# Add to end
fruits.append("cherry")  
print(fruits)  # ['apple', 'banana', 'cherry']

# Insert at position
fruits.insert(1, "orange")  
print(fruits)  # ['apple', 'orange', 'banana', 'cherry']

# Add multiple items
fruits.extend(["grape", "mango"])
print(fruits)  # ['apple', 'orange', 'banana', 'cherry', 'grape', 'mango']
```

### Removing Elements

```python
fruits = ["apple", "banana", "cherry", "date"]

# Remove by value
fruits.remove("banana")
print(fruits)
```

```python
fruits = ["apple", "banana", "cherry", "date"]

# Remove by index (default: last)
fruits.pop()
print(fruits)
```

```python
fruits = ["apple", "banana", "cherry", "date"]

# Remove by index
fruits.pop(1)
print(fruits)
```

```python
fruits = ["apple", "banana", "cherry", "date"]

# Remove all elements
fruits.clear()
print(fruits)   
```
