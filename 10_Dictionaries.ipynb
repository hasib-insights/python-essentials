### What is a Dictionary?
A dictionary is a mutable, unordered (Python 3.6+ ordered) collection 
of key-value pairs. Keys must be unique and immutable.

### Creating dictionaries

```python
empty_dict = {}
person = {
           "name" : "Hasib Hossain",
           "age" : 22 ,
           "country" : "Bangladesh",
           "is_student" : True

}

# Using dict() constructor
car = dict(brand="Toyota", 
           model="Corolla", 
           year=2020)
```

### Accessing Values

```python
person_info = {"name": "Hasib", "age" : 25, "country" : "Bangladesh" }

# Access by key
print(person_info["name"]) # Output: Hasib

# Using get() - safer
print(person_info.get("age")) # Output: 25
```

### Adding/Updating Elements

```python
person = {"name": "Hasib", "age": 25}

# Add new key-value pair
person["city"] = "Dhaka"
print(person)  
```

```python
# Update existing value
person["age"] = 26
print(person)  
```

```python
# Update multiple values
person.update({"job": "Engineer", "salary": 50000})
print(person) 
```

### Removing Elements

```python
person = {"name": "Hasib", "age": 25, "city": "Dhaka", "job": "Engineer"}

# Remove specific key
person.pop("age") 
print(person)   
```

```python
# Remove last inserted item
item = person.popitem()
print(item)  
```

```python
# Delete key
del person["city"]
print(person) 
```

```python
# Clear all
person.clear()
print(person)  
```
