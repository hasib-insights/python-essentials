### What is a Set?
A set is a mutable, unordered collection of unique elements.

```python
# Creating sets
empty_set = set()  # Note: {} is empty dictionary, not set!
numbers = {1, 2, 3, 4, 5}
fruits = {"apple", "banana", "cherry"}
```

```python
# Remove duplicates automatically
set = {1, 1, 2, 2, 3, 3, 4}
print(set) 
```

### Adding Elements

```python
fruits = {"apple", "banana"}

# Add single element
fruits.add("cherry")
print(fruits)  
```

```python
# Add multiple elements
fruits.update(["date", "elderberry"])
print(fruits)  
```

### Removing Elements

```python
fruits = {"apple", "banana", "cherry", "date"}

# Remove (raises error if not found)
fruits.remove("banana")
print(fruits)
```

```python
fruits = {"apple", "banana", "cherry", "date"}

# Remove and return random element
x =fruits.pop()
print(x)  # Random element
```

```python
# Clear all
fruits.clear()
print(fruits)  
```
