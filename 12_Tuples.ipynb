### What is a Tuple?
A tuple is an immutable (cannot change), ordered collection that allows duplicate elements.

### Creating tuples

```python
empty_tuple = ()
numbers = (1, 2, 3, 4, 5)
mixed = ("apple", 42, 3.14, True)
fruits = ("apple", "banana", "cherry")
```

```python
# Single element tuple (comma is important!)
single = (5,)  # Tuple
not_tuple = (5)  # Integer!
```

```python
# Without parentheses
point = 3, 4  # Tuple
```

### Accessing Elements

```python
fruits = ("apple", "banana", "cherry", "date")

print(fruits[0])    # Output: apple
print(fruits[-1])   # Output: date
print(fruits[1:3])  # Output: ('banana', 'cherry')
```

### Tuple Methods (Limited)

```python
numbers = (1, 2, 3, 2, 4, 2, 5)

print(len(numbers))       # 7
print(numbers.count(2))   # 3
print(numbers.index(2))   # 1 
```

### Tuple Unpacking

```python
# Unpack values
person = ("Hasib", 25, "Dhaka")
name, age, city = person
print(name)  # Hasib
print(age)   # 25
print(city)  # Dhaka
```

```python
# Ignore values with _
fruits = ("apple", "banana", "cherry")
fruit1, _, fruit3 = fruits
print(fruit1)  # apple
print(fruit3)  # cherry
```

```python
# Swapping variables
a, b = 5, 10
a, b = b, a
print(a)  # 10 
print(b)  # 5
```
