### Python Logical Operators
Logical operators are used to combine conditional statements. They work with **Boolean values** (`True`/`False`) and return a Boolean result.

| Operator | Name | Purpose | Syntax |
|----------|------|---------|--------|
| `and` | Logical AND | Returns `True` if **both** conditions are `True` | `x and y` |
| `or` | Logical OR | Returns `True` if **at least one** condition is `True` | `x or y` |
| `not` | Logical NOT | Reverses the Boolean value | `not x` |

---
### `and` – Truth Table

| A | B | A and B | Result |
|---|---|---------|--------|
| `True` | `True` | `True` |  True |
| `True` | `False` | `False` |  False |
| `False` | `True` | `False` |  False |
| `False` | `False` | `False` |  False |

### Quick Examples

| Condition | Code | Result |
|-----------|------|--------|
| Both true | `(5 > 3) and (2 < 4)` | `True` |
| First false | `(5 < 3) and (2 < 4)` | `False` |
| Second false | `(5 > 3) and (2 > 4)` | `False` |
| Both false | `(5 < 3) and (2 > 4)` | `False` |

### Check Voting Eligibility

```python
age = 20
is_citizen = True

if age >= 18 and is_citizen:
    print("You can vote")
else:
    print("You can't vote")    
```

___
### `or` – Truth Table

| A | B | A or B | Result |
|---|---|--------|--------|
| `True` | `True` | `True` |  True |
| `True` | `False` | `True` | True |
| `False` | `True` | `True` | True |
| `False` | `False` | `False` | False |

### Quick Examples

| Condition | Code | Result |
|-----------|------|--------|
| Both true | `(5 > 3) or (2 < 4)` | `True` |
| First true | `(5 > 3) or (2 > 4)` | `True` |
| Second true | `(5 < 3) or (2 < 4)` | `True` |
| Both false | `(5 < 3) or (2 > 4)` | `False` |

### Weekend Check

```python
day = "Monday"

if day == "Friday" or day == "Saturday":
    print("It's the weekend!")
else:
    print("It's a working day!")  
```

---
### `not` – Truth Table

| A | not A |
|---|-------|
| `True` | `False` |
| `False` | `True` | 

### Quick Examples

| Condition | Code | Result |
|-----------|------|--------|
| Not true | `not (5 > 3)` | `False` |
| Not false | `not (5 < 3)` | `True` |
| Not equal | `not x == 5` | Equivalent to `x != 5` |
| Not empty | `not my_list` | `True` if empty |

### Weather Check

```python
is_raining = True

if not is_raining:
    print("You can go outside")
else:
    print("Take an umbrella!")    
```

### Membership Operators 
Membership operators are used to test if a value exists in a **sequence** (string, list, tuple, set, dictionary) or **collection**. They return `True` if the value is found, otherwise `False`.

| Operator | Example | Result |
|----------|---------|--------|
| `in` | `"apple" in ["apple", "banana"]` | `True` |
| `in` | `"grape" in ["apple", "banana"]` | `False` |
| `not in` | `"grape" not in ["apple", "banana"]` | `True` |
| `not in` | `"apple" not in ["apple", "banana"]` | `False` |

### Check if character/substring  exists in string

```python
text = "Hello World"

print("H" in text)    # True
print("e" in text)    # True
print("x" in text)    # False

print("World" in text)   # True
print("world" in text)   # False (case-sensitive)
print("Hello" in text)   # True
```

### Identity Operators
Identity operators are used to compare the **memory location** of two objects. They check if two variables point to the **same object** in memory, not if they have the same value.
| Operator | Name | Example | 
|----------|---------|--------|
| is       | identity| x is y |
| is not   | not identity | x is not y |


```python
age_1 = 20
age_2 = 25

age_1 is age_2       # False
age_1 is not age_2   # True
```

### Bitwise Operators
Bitwise operators work on **bits** (binary digits) and perform operations at the bit level. They are used on integers and return integer results.

| Operator | Name | Purpose |Example|
|----------|------|---------|-------|
| `&` | AND | Sets each bit to 1 if **both** bits are 1 |`x & y`|
| `\|` | OR | Sets each bit to 1 if **at least one** bit is 1 |`x \| y`| 
| `^` | XOR | Sets each bit to 1 if **only one** bit is 1 |`x ^ y`|
| `~` | NOT | Inverts all the bits |`~ x`|
| `<<` | Left Shift | Shifts bits left, fills with 0 |`x << y`|
| `>>` | Right Shift | Shifts bits right |`x >> y`|
