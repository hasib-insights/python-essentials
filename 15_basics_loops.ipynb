###  Python Loops 
Loops are used to **repeat a block of code** multiple times. They are essential for automating repetitive tasks.
| Loop Type | When to Use |
|-----------|-------------|
| `for` loop | When you know how many times to repeat |
| `while` loop| When you don't know how many times to repeat |

> **Think of it like:** A loop is like a washing machine cycle – it repeats the same process until the task is complete.

### FOR Loop

````python
#Looping Through a List

fruits = ["apple", "banana", "cherry", "date"]

for fruit in fruits:         # for {variable} in {list_name}:
    print(f"I like {fruit}")
````

````python
# Looping Through a String
word = "Python"

for letter in word:
    print(letter)
````

````python
# Using range(stop)
for i in range(5): # 0 - 4
    print(i)
````

````python
#  Using range(start,stop)
for i in range(2,6): # 2 - 5
    print(i)
````

````python
# using range(start,stop,step)
for i in range(1,10,2): # 1,3,5,7,9
    print(i)
````

````python
# sing enumerate() (Get Index + Value)
fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits):
    print(f"{index}. {fruit}")
````

### Nested For Loop 
>Loop inside a Loop

````python
# Multiplication table

for multiplicand in range(1, 4):
    for multiplier in range(1, 4):
        print(f"{multiplicand} x {multiplier} = {multiplicand * multiplier}")
    print("-" * 20)  # separate tables with a line of dashes  
````

````python
# Print numbers 0 to 4

i = 0
while i < 5:
    print(i)
    i += 1   # increment i by 1
````

````python
# Countdown from 5 to 1 
countdown = 5

while countdown > 0:
    print(countdown)
    countdown -= 1

````

````python
# calculate the sum of numbers from 1 to 10

total = 0
i = 1
while i <= 10:  # loop until i is less than or equal to 10
    total += i  # add i to total
    i += 1      # increment i by 1
    
print(f"sum = {total}")
````
